```python
import sklearn
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import math
import pickle

def read_file(str_file):
    f = open(str_file, 'r')
    x = f.readlines()
    x = [float(y) for y in x]
    return x

def read_gamma(str_file):
    f = open(str_file, 'r')
    x = f.readlines()
    x = [str(y) for y in x]
    return x


comp_list = read_file('Dev_results/comp_list.txt')
c_list = read_file('Dev_results/c_list.txt')
vocab_list = read_file('Dev_results/vocab_list.txt')
tfidf_list = read_file('Dev_results/tfidf_list.txt')
gamma_list = read_gamma('Dev_results/gamma_list.txt')


acc_list_comp = read_file('Dev_results/acc_list_comp.txt')
acc_list_c = read_file('Dev_results/acc_list_c.txt')
acc_list_voc = read_file('Dev_results/acc_list_voc.txt')
acc_list_tf = read_file('Dev_results/acc_list_tf.txt')
acc_list_gamma = read_file('Dev_results/acc_list_gamma.txt')

t_learn_list_comp = read_file('Dev_results/t_learn_list_comp.txt')
t_learn_list_c = read_file('Dev_results/t_learn_list_c.txt')
t_learn_list_voc = read_file('Dev_results/t_learn_list_voc.txt')
t_vectorize_list_voc = read_file('Dev_results/t_vectorize_list_voc.txt')
t_learn_list_tf = read_file('Dev_results/t_learn_list_tf.txt')
t_vectorize_list_tf = read_file('Dev_results/t_vectorize_list_tf.txt')
t_learn_list_gamma = read_file('Dev_results/t_learn_list_gamma.txt')

t_transform_voc = read_file('Dev_results/t_transform_voc.txt')
t_transform_tf = read_file('Dev_results/t_transform_tf.txt')
```

```python
lin = sklearn.linear_model.LinearRegression()
lin.fit(np.asarray(vocab_list).reshape(-1,1), np.asarray(t_vectorize_list_voc).reshape(-1,1))
print(lin.coef_)
plt.plot(vocab_list, t_vectorize_list_voc)
plt.title("Word frequency vector feature size against time to preprocess")
plt.ylabel("Time/s")
plt.xlabel("Feature size")
plt.show()
plt.plot(vocab_list, t_learn_list_voc)
plt.title("Word frequency vector feature size against time to train SVM")
plt.ylabel("Time/s")
plt.xlabel("Feature size")
plt.show()
plt.plot(vocab_list, acc_list_voc)
plt.title("Word frequency vector feature size against accuracy")
plt.ylabel("Accuracy score")
plt.xlabel("Feature size")
plt.show()

x = zip(vocab_list, acc_list_voc)
for z in x:
    print(z)
```

```python
print(comp_list)
x = (zip(acc_list_comp, t_learn_list_comp))
for y in x:
    print(y)
    

plt.plot(t_learn_list_comp, acc_list_comp)
plt.title("Time to train SVM against accuracy score")
plt.ylabel("Accuracy score")
plt.xlabel("Time/s")
plt.show()

plt.plot(comp_list, acc_list_comp)
plt.title("PCA components against accuracy score")
plt.ylabel("Accuracy score")
plt.xlabel("PCA components")
plt.show()

from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=2, include_bias=False)  # create a 2nd degree feature
X_poly = poly_features.fit_transform(np.asarray(t_learn_list_comp).reshape(-1,1))
plt.plot([1,2,3,4,5,6,7],X_poly)
plt.show()
```

```python
print(c_list)
plt.plot(c_list, acc_list_c)
plt.title("C regularization parameter against accuracy score")
plt.ylabel("Accuracy score")
plt.xlabel("C parameter")
plt.show()

plt.plot(c_list, t_learn_list_c)
plt.title("C regularization parameter against time to train SVM")
plt.ylabel("Time/s")
plt.xlabel("C parameter")
plt.show()

x = zip(c_list, t_learn_list_c)
for z in x:
    print(z)
```

```python
plt.plot(tfidf_list, acc_list_tf)
plt.title("TF-IDF feature size against accuracy score")
plt.ylabel("Accuracy score")
plt.xlabel("TF-IDF feature size")
plt.show()

plt.plot(tfidf_list, t_learn_list_tf)
plt.title("TF-IDF feature size against time to train SVM")
plt.ylabel("Time/s")
plt.xlabel("TF-IDF feature size")
plt.show()

plt.plot(tfidf_list, t_vectorize_list_tf)
plt.title("TF-IDF feature size against time to preprocess")
plt.ylabel("Time/s")
plt.xlabel("TF-IDF feature size")
plt.show()
```

```python
plt.plot(gamma_list, acc_list_gamma)
plt.title("Gamma value against accuracy score")
plt.xlabel("Gamma")
plt.ylabel("Accuracy score")
plt.show()

plt.plot(gamma_list, t_learn_list_gamma)
plt.title("Gamma value against time to train SVM")
plt.xlabel("Gamma")
plt.ylabel("Time/s")
plt.show()
```
